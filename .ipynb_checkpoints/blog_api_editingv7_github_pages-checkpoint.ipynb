```
def generate_blog(topic):
    model = genai.GenerativeModel("gemini-flash-latest")
    prompt = f"""Write a structured blog post with title, headings, and conclusion.
Topic: {topic}"""
    return model.generate_content(prompt).text
```

```
def save_blog_post(title, content):
    date = datetime.date.today().isoformat()
    filename = f"_posts/{date}-{title.replace(' ', '-').lower()}.md"
    front_matter = f"""---
layout: post
title: "{title}"
date: {date}
---
"""
    with open(filename, "w", encoding="utf-8") as f:
        f.write(front_matter + "\n" + content)
    print(f"Saved {filename}")
```

```
TOPIC = "Why the Dodo cannot fly?"
content = generate_blog(TOPIC)
save_blog_post(TOPIC, content)
```

```
import datetime
import re
import os
```

```
def save_blog_post(title, content):
    date = datetime.date.today().isoformat()   # YYYY-MM-DD
    slug = re.sub(r'[^a-z0-9]+', '-', title.lower()).strip('-')

    filename = f"{date}-{slug}.md"

    # Pre-process the title to handle quotes before embedding in the f-string
    cleaned_title = title.replace('"', "'")

    front_matter = f"""---
layout: post
title: "{cleaned_title}"
date: {date}
---
"""

    with open(filename, "w", encoding="utf-8") as f:
        f.write(front_matter + "\n" + content)

    print(f"✅ Saved blog post as: {filename}")
    return filename
```

```
saved_filename = save_blog_post(TOPIC, content)
print(f"Blog post saved as: {saved_filename}")
```
